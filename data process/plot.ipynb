```python
import pandas as pd

df = pd.read_csv(r"E:\Sort Data\cham4_mix\cham4_4variable_4.csv")

# Step 1: Calculate the mean intensity value (excluding values greater than 10000 and "fl6" entries)
mean_intensity = df[(df['intensity'] <= 10000) & (df['id_floor'] != 'fl6')]['intensity'].mean()
print(mean_intensity)

# Step 2: Update the intensity values greater than 10000 (and not in "fl6") with the calculated mean
df.loc[(df['intensity'] > 10000) & (df['id_floor'] != 'fl6'), 'intensity'] = mean_intensity

df.to_csv(r"E:\Sort Data\cham4_mix\cham4_4variable_4.csv", index=False)


```

```python
import matplotlib.pyplot as plt
import pandas as pd
 
df = pd.read_csv(r"E:\Sort Data\cham4_time_format_test.csv")

# Convert the 'time' column to datetime objects
df['time'] = pd.to_datetime(df['time'])

# Filter only 'fl1' data
fl_df = df[df['id_group'] == 'fl2']

# Plot the energy values against time for 'fl1'
plt.figure(figsize=(24, 6))
plt.plot(fl_df['time'], fl_df['value'], marker='o', linestyle='-')
plt.xlabel('Time')
plt.ylabel('Energy')
plt.title('Energy vs. Time (fl2)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Save the plot as a .jpg file
plt.savefig('E:\Sort Data\graph\cham4\energy_fl2_plot.jpg', dpi=300)

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv(r"E:\Sort Data\cham4_mix\cham4_4variable_2.csv", parse_dates=['time'])

def plot(floor_number, variable):

    # Filter only 'fl1' data
    fl_df = df[df['id_floor'] == f'fl{floor_number}']

    # Plot the energy values against time for 'fl1'
    plt.figure(figsize=(24, 6))
    plt.plot(fl_df['time'], fl_df[variable], marker='o', linestyle='-')
    plt.xlabel('Time')
    plt.ylabel(variable)
    plt.title(f'{variable} vs. Time (fl{floor_number})')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()

    # Save the plot as a .jpg file
    plt.savefig(f'E:\\Sort Data\\test\\{variable}_fl{floor_number}_plot2in.jpg', dpi=300)

    plt.show()

#energy / temperature / humidity / intensity
variable = 'energy' 

for i in range(df['id_floor'].nunique()):
    plot(i+1, variable)
```
